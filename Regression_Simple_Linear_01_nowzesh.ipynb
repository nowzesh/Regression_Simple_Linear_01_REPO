```python
# Importing python libraries 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the Dataset from a CSV File 
dataset = pd.read_csv('Salary_Data.csv')

# View 1st five rows 
dataset.head()
```

```python
# Feature Extraction 

# Independent features 
x = dataset.iloc[:, 0:-1].values # all rows, all columns except the last column (last column: target variable)
print(f"\nIndependent variables, x: \n{x}")

y = dataset.iloc[:, -1:] # all rows, last column, 2x2 array 
print(f"\nDependable variable, y: \n{y}")
print(f"Type of array, y: {type(y)}")
print(f"Type of array dimension, y: {y.ndim}") # returns 1D or 2D array 
print(f"Type of array shape, y: {y.shape}") # returns row number and column numbers  
```

```python
# check missing values columnwise 
missing_values = dataset.isnull().sum()
print(missing_values)
```

```python
# Split independent & dependent variables into train & test dataset 

# import python libraries 
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 1/3, random_state= 0)

print(f"\nTraining set: (independent) \n{x_train}\nTraining set: (dependent) \n{y_train}\nTest set: (independent) \n{x_test}\nTest set: (dependent) \n{y_test}")
```

```python
# Simple Linear Regression Model 

# Import Regresison Library 
from sklearn.linear_model import LinearRegression 

regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Prediction from test data set using Linear Regression 

y_pred = regressor.predict(x_test)

print(y_pred)
print(y_pred.shape)

plt.plot(x_test, y_pred, color = 'blue', label='Predicted Salary')
plt.scatter(x_train, y_train, color = 'red', label='Actual Salary')
plt.title("Actual vs Predicted Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
